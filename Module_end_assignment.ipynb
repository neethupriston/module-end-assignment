```python
Determine the distribution of employees across each team and calculate 
the percentage split relative to the total number of employees.
```

```python
team_counts = df['Team'].value_counts()
team_counts
```

```python
team_percentages = round((team_counts / df.shape[0]) * 100, 2)
team_percentages
```

```python
team_distribution = pd.DataFrame({
    'Employee Count': team_counts,
    'Percentage (%)': team_percentages})
team_distribution
```

```python
plt.figure(figsize=(10, 6))
plt.bar(team_counts.index, team_percentages )
plt.xlabel("Team")
plt.ylabel("Percentage of Employees")
plt.title("Distribution of Employees Across Teams")
plt.xticks(rotation=45, ha="right")  
plt.tight_layout() 
plt.show()
```

New Orleans Pelicans is the Team with highest number of employees and Minnesota Timberwolves is the Team with lowest number of employees

Discover which team and position have the highest salary expenditure. 

```python
plt.figure(figsize=(10, 6))
sns.scatterplot(x='Team', y='Salary', hue='Position', size='Salary', data=df)
plt.xticks(rotation=60, ha="right")
plt.xlabel('Team')
plt.ylabel('Total Salary Expenditure')
plt.title('Salary Expenditure by Team and Position')
plt.tight_layout()
plt.show()
```

SF is the position and Los Angeles Lakers is the team with highest salary expenditure
